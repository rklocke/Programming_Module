#### 1.3. Given two positive integers a and b, return the sum of all odd integers from a to b inclusively.
So if a = 50 and b = 100 the output = 1875
Challenge: what is the output if a = 10 and b = 25?

```python
#Becky
def add_odds(a,b):
    totalodds=0
    for i in range(a,b+1):
        if i%2!=0:
            totalodds+=i
    return totalodds

def total_odds(a,b):
    return sum([x for x in range(a,b+1) if x%2!=0])

print(add_odds(10,25))
print(total_odds(10,25))
```
